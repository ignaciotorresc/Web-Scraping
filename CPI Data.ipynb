```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
data = requests.get('https://www.bls.gov/news.release/cpi.nr0.htm').text
```

```python
soup = BeautifulSoup(data, 'html.parser')
```

```python
headers=[]

for i in soup.find('thead').find_all('tr')[1].find_all('th'):
    title= i.text
    headers.append(title)

headers
```

```python
final_column=[]

for i in soup.find('thead').find_all('tr')[0].find_all('th')[-1]:
    column= i.text
    final_column.append(column)

final_column=''.join(final_column)

final_column
```

```python
headers.append(final_column)
headers
```

```python
df=pd.DataFrame(columns=headers,)
df
```

```python
for row in soup.find('tbody').find_all('tr'):
    data = row.find_all('td')
    row_data = [td.text.strip() for td in data]
    lenght= len(df)
    df.loc[lenght]=row_data
```

```python
df
```

```python
items=[]

for i in soup.find('tbody').find_all('tr'):
    itemss= i.find('th')
    items.append(itemss.text)
    
items
```

```python
df['']=items
```
